```python
# Charger une image depuis votre ordinateur vers le notebook
from google.colab import files
uploaded = files.upload()
image_path = list(uploaded.keys())[0]
```

```python
# Lire l'image en couleur
image_color = cv2.imread(image_path)
```

```python
# Convertir l'image en niveaux de gris
image_gray = cv2.cvtColor(image_color, cv2.COLOR_BGR2GRAY)
```

```python
# Afficher l'image en niveaux de gris
plt.imshow(image_gray, cmap='gray')
plt.title('Image en niveaux de gris')
plt.axis('off')
plt.show()
```

```python
import numpy as np
```

```python
# Calculer la PDF (fonction de densité de probabilité)
pdf, bins = np.histogram(image_gray.flatten(), bins=255, range=[0,255], density=True)
```

```python
# Calculer la CDF (fonction de distribution cumulative)
cdf = np.cumsum(pdf)
```

```python
# Égaliser l'histogramme en utilisant la CDF
image_equalized = np.interp(image_gray.flatten(), bins[:-1], cdf * 255).reshape(image_gray.shape).astype(np.uint8)
```

```python
# Afficher l'image originale et l'image égalisée
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB))
plt.title('Image Originale')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(image_gray, cmap='gray')
plt.title('Image en niveaux de gris')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(image_equalized, cmap='gray')
plt.title('Image Égalisée')
plt.axis('off')

plt.show()



```

```python
# Afficher les histogrammes côte à côte
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.hist(image_gray.ravel(), bins=255, range=[0, 255], density=True, color='b', alpha=0.75, label='Image en niveaux de gris',log=True)
plt.title('Histogramme de l\'image en niveaux de gris')
plt.xlabel('Niveau de gris')
plt.ylabel('Fréquence')
plt.legend()

plt.subplot(1, 2, 2)
plt.hist(image_equalized.ravel(), bins=255, range=[0, 255], density=True, color='g', alpha=0.75, label='Image Égalisée',log=True)
plt.title('Histogramme de l\'image égalisée')
plt.xlabel('Niveau de gris')
plt.ylabel('Fréquence')
plt.legend()
plt.show()
```

# Comparaison des deux histogrammes
L'égalisation de l'histogramme a  contribué à améliorer la répartition des niveaux de gris dans l'image, conduisant à un meilleur contraste et potentiellement à une meilleure visibilité des détails. La comparaison des histogrammes offre des insights visuels sur ces améliorations.

## Comparaison des deux images
L'image égalisée présente une amélioration significative du contraste, rendant les détails plus saillants et les contours plus nets. Les niveaux de gris sont répartis de manière plus uniforme, éliminant les zones de faible contraste de l'image en niveaux de gris d'origine. Cette égalisation renforce la clarté visuelle et la lisibilité des caractéristiques de l'image.

```python
#AFFICHAGE DES MATRICED DE L'IMAGE EN GRIS ET L'IMAGE L'ÉGALISÉ

# Afficher les matrices des images en niveaux de gris et égalisée
print("Matrice de l'image en niveaux de gris :")
print(image_gray)

print("\nMatrice de l'image égalisée :")
print(image_equalized)
```
