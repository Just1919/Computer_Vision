```python
# Charger une image depuis votre ordinateur vers le notebook
from google.colab import files
uploaded = files.upload()
image_path = list(uploaded.keys())[0]
```

```python
# Lire l'image en couleur
image_color = cv2.imread(image_path)
```

```python
# Convertir l'image en niveaux de gris
image_gray = cv2.cvtColor(image_color, cv2.COLOR_BGR2GRAY)
```

```python
# Afficher l'image en niveaux de gris
plt.imshow(image_gray, cmap='gray')
plt.title('Image en niveaux de gris')
plt.axis('off')
plt.show()
```

```python
import numpy as np
```

```python
# Calculer la PDF (fonction de densit de probabilit)
pdf, bins = np.histogram(image_gray.flatten(), bins=255, range=[0,255], density=True)
```

```python
# Calculer la CDF (fonction de distribution cumulative)
cdf = np.cumsum(pdf)
```

```python
# galiser l'histogramme en utilisant la CDF
image_equalized = np.interp(image_gray.flatten(), bins[:-1], cdf * 255).reshape(image_gray.shape).astype(np.uint8)
```

```python
# Afficher l'image originale et l'image galise
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB))
plt.title('Image Originale')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(image_gray, cmap='gray')
plt.title('Image en niveaux de gris')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(image_equalized, cmap='gray')
plt.title('Image galise')
plt.axis('off')

plt.show()



```

```python
# Afficher les histogrammes cte  cte
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.hist(image_gray.ravel(), bins=255, range=[0, 255], density=True, color='b', alpha=0.75, label='Image en niveaux de gris',log=True)
plt.title('Histogramme de l\'image en niveaux de gris')
plt.xlabel('Niveau de gris')
plt.ylabel('Frquence')
plt.legend()

plt.subplot(1, 2, 2)
plt.hist(image_equalized.ravel(), bins=255, range=[0, 255], density=True, color='g', alpha=0.75, label='Image galise',log=True)
plt.title('Histogramme de l\'image galise')
plt.xlabel('Niveau de gris')
plt.ylabel('Frquence')
plt.legend()
plt.show()
```

# Comparaison des deux histogrammes
L'galisation de l'histogramme a  contribu  amliorer la rpartition des niveaux de gris dans l'image, conduisant  un meilleur contraste et potentiellement  une meilleure visibilit des dtails. La comparaison des histogrammes offre des insights visuels sur ces amliorations.

## Comparaison des deux images
L'image galise prsente une amlioration significative du contraste, rendant les dtails plus saillants et les contours plus nets. Les niveaux de gris sont rpartis de manire plus uniforme, liminant les zones de faible contraste de l'image en niveaux de gris d'origine. Cette galisation renforce la clart visuelle et la lisibilit des caractristiques de l'image.

```python
#AFFICHAGE DES MATRICED DE L'IMAGE EN GRIS ET L'IMAGE L'GALIS

# Afficher les matrices des images en niveaux de gris et galise
print("Matrice de l'image en niveaux de gris :")
print(image_gray)

print("\nMatrice de l'image galise :")
print(image_equalized)
```
